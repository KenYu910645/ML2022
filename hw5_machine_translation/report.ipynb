```python
import pandas as pd 
from matplotlib import pyplot as plt 
import numpy as np
df = pd.read_csv("p1.csv")
print(np.array(df['gnorm']))
plt.figure(figsize=(20,10))
plt.plot(np.array(df['t']), np.array(df['gnorm']))
plt.title("gnorm")
plt.show()
```

```python
import pickle
import numpy as np, matplotlib, matplotlib.pyplot as plt, seaborn as sns

with open("p2.pkl", "rb") as f:
    sim_mat = pickle.load(f)
f, ax = plt.subplots(figsize=(20, 20))
sns.set_theme(style = "white")
sns.heatmap(sim_mat, vmin=0, vmax=1,
            square=True, xticklabels = False, yticklabels = False)
plt.show()
```
